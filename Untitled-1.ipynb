```python
lons = single_pressure.coord("longitude").points
lats = single_pressure.coord("latitude").points

fig, ax = plt.subplots()

cntr = ax.contourf(lons, lats, single_pressure.data)
cb = fig.colorbar(cntr)
cb.ax.set_title(f'[{single_pressure.units}]')

ax.set_xlabel('Longitude [$\degree$]')
ax.set_ylabel('Latitude [$\degree$]')
ax.set_title(
    f"{single_pressure.name().capitalize().replace('_', ' ')}  ",
);
```

```python
height_10 = t_pres[:,40,:,:]

#making a movie
def plotter(height, time):
    height = t_pres[:,height,:,:]
    for t in range(0,time):
        time_slice = height[t]
        lons = time_slice.coord("longitude").points
        lats = time_slice.coord("latitude").points

        fig, ax = plt.subplots()

        cntr = ax.contourf(lons, lats, time_slice.data)
        cb = fig.colorbar(cntr)
        cb.ax.set_title(f'[{time_slice.units}]')

        ax.set_xlabel('Longitude [$\degree$]')
        ax.set_ylabel('Latitude [$\degree$]')
        ax.set_title(
        f"{time_slice.name().capitalize().replace('_', ' ')}  ",);
        timename = "%s.png" %t
        plt.savefig(timename)
        
         
    
```

```python
plotter(5,13)
```

```python
def plotter_height(height, time):
    time = t_pres[time,:,:,:]
    for h in range(0,height):
        height_slice = time[h]
        lons = height_slice.coord("longitude").points
        lats = height_slice.coord("latitude").points

        fig, ax = plt.subplots()

        cntr = ax.contourf(lons, lats, height_slice.data)
        cb = fig.colorbar(cntr)
        cb.ax.set_title(f'[{height_slice.units}]')

        ax.set_xlabel('Longitude [$\degree$]')
        ax.set_ylabel('Latitude [$\degree$]')
        ax.set_title(
        f"{height_slice.name().capitalize().replace('_', ' ')}  ",);
        timename = "%s.png" %h
        plt.savefig(timename)
```

```python
plotter_height(67, 10)
```
