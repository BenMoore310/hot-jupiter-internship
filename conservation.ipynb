```python
#equilibrium, 10x solar


# parents_dir = Path.cwd().parent.parent.parent.parent/"data" / "mz355" /"um_runs"/"wasp96b"/"equilibrium"/"u-cp770"/"raw"
parents_dir_eq1 = Path("/data/mz355/um_runs/wasp96b/equilibrium/u-cp770/raw")
flist = []

# print(parents_dir)
df_eq1 = pd.DataFrame()

for data_file_eq1 in parents_dir_eq1.glob('**/atmos_base_*/conservation_diag.dat'):  # this directory and all subdirectories, recursively

    # f=open(data_file, "r")
    # lines = f.readlines()
    # for x in lines:
    #     flist.append(x.split()[0])

    # f.close()
    df_eq1 = pd.concat([df_eq1, pd.read_table(data_file_eq1, header=None, sep='   ', names=["timestep", "mass", "angular_momentum", "KE"], engine="python")])
    


df_eq1.sort_values("timestep", inplace=True)
print(df_eq1.head())
df_eq1["mass"] = df_eq1["mass"].apply(lambda x: x / 2.410648e+23)
df_eq1["angular_momentum"] = df_eq1["angular_momentum"].apply(lambda x: x / 2.446091e+34)
df_eq1["KE"] = df_eq1["KE"].apply(lambda x: x / 9.950315e+23)
df_eq1["timestep"] = df_eq1["timestep"].apply(lambda x: x/2880)
# for label in zip(["mass", "angular_momentum", "KE"]):
#     df[label] = df[label]

fig, axs = plt.subplots(1,3, figsize=(24,6))
fig.suptitle("10x solar metallicity")
for label, ax in zip(["mass", "angular_momentum", "KE"], axs.reshape(-1)):
    
    # ax.set_title("10x solar metallicity")
    ax.plot(df_eq1["timestep"], df_eq1[label], linewidth=1)
    ax.set_ylabel(label)
    ax.set_xlabel("timestep / days")
    
    

```

```python
#Horizontal winds, eq1
tgt_plevs = np.logspace(7, 2, 65)  # target pressure levels
from aeolus.coord import interp_cube_from_height_to_pressure_levels


data_file = parents_dir_eq1 /"atmos_base_0090" / "atmosa.pb00000080_00"
pt = iris.load(data_file)
print(pt)
data_file_2 = parents_dir_eq1 /"atmos_base_0090" / "atmosa.pc00000080_00"
winds = iris.load(data_file_2)
winds
```

```python
tgt_plevs
```

```python
u = winds.extract_cube("x_wind")
v = winds.extract_cube("y_wind")
print(u.shape, v.shape)
```

```python
lons = u.coord("longitude").points
lats = u.coord("latitude").points
```

```python
pressure = pt.extract_cube("air_pressure")
pressure = pressure[0:66,:,:]
v = v.regrid(u, iris.analysis.Linear())

print(pressure.shape)
```

```python
print(pressure)
```

```python
u_plevs = interp_cube_from_height_to_pressure_levels(u, pressure, tgt_plevs)
v_plevs = interp_cube_from_height_to_pressure_levels(v, pressure, tgt_plevs)
print(u_plevs.shape, v_plevs.shape)
```

```python
# u2d = u_plevs.extract(iris.Constraint(level_height=1e4))
# v2d = v_plevs.extract(iris.Constraint(level_height=1e4))

print(v_plevs[51])
```

```python
# print(u2d.coord("level_height"))

u_plevs_e3 = u_plevs[51]
u_plevs_e4 = u_plevs[38]
u_plevs_e5 = u_plevs[25]

v_plevs_e3 = v_plevs[51]
v_plevs_e4 = v_plevs[38]
v_plevs_e5 = v_plevs[25]
```

```python
wspd = (u_plevs_e4**2 + v_plevs_e4**2) ** 0.5
print(wspd)
fig, ax = plt.subplots(figsize = (15,10))

lon_slice = slice(None, None, 4)  # every 4th point in longitude
lat_slice = slice(None, None, 4)  # every 5th point in longitude

quiv = ax.quiver(
    lons[lon_slice],
    lats[lat_slice],
    u_plevs_e4.data[lat_slice, lon_slice],
    v_plevs_e4.data[lat_slice, lon_slice],
    wspd.data[lat_slice, lon_slice],
    cmap="magma_r",
   
)
ax.set_title("")
ax.set_ylabel("Longitude")
ax.set_xlabel("latitude")
ax.set_title("Horizontal wind velocity, 1e4 Pa, Equilibrium MdH1")

cb = fig.colorbar(quiv)
cb.ax.set_ylabel(f"[{wspd.units}]");
```
